```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import datetime as dt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from scipy.cluster.hierarchy import linkage
from scipy.cluster.hierarchy import dendrogram
from scipy.cluster.hierarchy import cut_tree
```

```python
# read the dataset
cricket = pd.read_csv("Cricket.csv", sep=",", encoding="ISO-8859-1", header=0)
cricket.head()
```

```python
new_df=cricket[['SR','Ave']]
```

```python
new_df.head()
```

```python
scaler=StandardScaler()
new_df=pd.DataFrame(scaler.fit_transform(new_df))
new_df.head()
```

```python
kmeans=KMeans(n_clusters=4,max_iter=50,random_state=100)
kmeans.fit(new_df)
```

```python
kmeans.labels_
```

```python
new_df['label']=kmeans.labels_
new_df['name']=cricket.Player
```

```python
new_df[new_df['label']==3]
```

```python
new_df.drop(columns=['label','name'],inplace=True)
```

```python
mergings=linkage(new_df,method='complete',metric='euclidean')
dendrogram(mergings)
plt.show()
```

```python
cluster_labels = cut_tree(mergings,n_clusters=3).reshape(-1,)
cluster_labels
```

```python
new_df['label']=cluster_labels
new_df['name']=cricket.Player
```
